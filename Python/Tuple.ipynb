```python
my_tuple = (1, 2, 3)
my_list = [1, 2, 3]
```

```python
my_list[0] = 4  # The first element of my_list is now 4
# my_tuple[0] = 4  # This will raise a TypeError: 'tuple' object does not support item assignment

```

```python
my_list.append(4)  # Adds 4 to the end of my_list

# This cannot be done with a tuple because tuples are immutable
```

```python
my_list.remove(2)  # Removes the first occurrence of 2 from my_list

# This cannot be done with a tuple because tuples are immutable
```

```python
my_tuple = (1, 2, 3)
my_other_tuple = (4, 5, 6)
new_tuple = my_tuple + my_other_tuple

my_list = [1, 2, 3]
my_other_list = [4, 5, 6]
new_list = my_list + my_other_list
```
