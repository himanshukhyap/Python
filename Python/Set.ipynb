```python
# A set is a collection which is unordered(isme koi order se jame nahi rahte hai), 
# unchangeable (change nahi kar sakte hai lekin add and remove kar sakte hai )*, 
# and unindexed ( sets are unindexed, which means that you cannot access elements of the set by index. ).
```

```python
# Duplicates Not Allowed
thisset = {"apple", "banana", "cherry", "apple",1,2,5,10,0}

print(thisset)
```

```python
# Create a set of unique elements
my_set = {1, 2, 3, 4, 5}

# Create a list of elements that may contain duplicates
my_list = [1, 2, 3, 4, 5, 5]
```

```python
# True and 1 is considered the same value:

thisset = {"apple", "banana", "cherry", True, 1, 2}

print(thisset)
```

```python
# Adding elements to a set
my_set.add(6)

# Adding elements to a list
my_list.append(6)
```

```python
# Removing elements from a set
my_set.remove(6)

# Removing elements from a list
my_list.remove(6)
```

```python
colors = ["red", "green", "blue", "red", "orange", "green"]

# Remove duplicates using a set
unique_colors = set(colors)
unique_colors
```

```python
# Sort the remaining elements using a list
sorted_colors = sorted(unique_colors)

print(sorted_colors) # Output: ["blue", "green", "orange", "red"]
```
