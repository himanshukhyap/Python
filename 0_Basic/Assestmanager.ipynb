```python
import pymongo
import requests
import json
```

```python
coingecko = 'https://api.coingecko.com/api/v3/coins/list?include_platform=false'
coincap =   'https://api.coincap.io/v2/assets/bitcoin'
```

```python

def AddAssestManagerData(obj):
    myclient = pymongo.MongoClient("mongodb://localhost:27017")
    mydb = myclient["AssestCoin"]
    mycol = mydb["Api_assestmanager"]
    mycol.insert_one(obj)
```

```python
def jprint(obj):
    # create a formatted string of the Python JSON object
    text = json.dumps(obj, sort_keys=False, indent=2)
    print(text)
```

```python
def SingleGetData(api):
        response = requests.get(api)
        if response.status_code == 200:
             obj = {'AssestId':response.json()['id'], 'Symbol': response.json()['symbol'], 'Name': response.json()['name'] , 'Image':response.json()['image']  }             
            # objectModel(response.json()['id'],response.json()['symbol'],response.json()['name'],response.json()['image'] )
             AddAssestManagerData(obj)
             print(obj)
        else:
            print("error SingleGetData")
```

```python
def getAssestManagerData(api):
        response = requests.get(f"{api}")
        # print(response.status_code)
        # print(response.json())
        # print(response.text)
        if response.status_code == 200:
            for x in response.json(): 
                url= 'https://api.coingecko.com/api/v3/coins/{}'.format(x['id'])
                SingleGetData(url)            
        else:
            print("error")

```
