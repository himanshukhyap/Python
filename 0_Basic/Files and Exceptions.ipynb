```python
import pathlib as p

path_ = p.Path('/content/ pi_digits.txt ')
contents = path_.read_text()
# We can remove the extra blank line by using rstrip() on the contents string:
# rstrip() stands for "right strip," indicating that it removes characters from the right end (or "tail") of a string.
contents = contents.rstrip()
print(contents)
```

```python
text = "   Hello, World!   "
print(text.rstrip())  # Output: "   Hello, World!"
```

```python
lines = contents.splitlines()
for line in lines:
    print(line.rstrip())
```
