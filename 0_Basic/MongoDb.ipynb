```python
import json
def jprint(obj):
    print(json.dumps(obj,sort_keys=False, indent=2))
```

```python
import requests
import json

response = requests.get('https://dummyapi.online/api/movies?id=1')
if response.status_code == 200:
    data = json.loads(response.text)
    jprint(data)
    
```

```python
from urllib.parse import urlparse

url = 'https://dummyapi.online/api/movies?id=1'
parsed_url = urlparse(url)
print(parsed_url.scheme)   # Output: 'http'
print(parsed_url.netloc)   # Output: 'dummyapi.online'
print(parsed_url.path)     # Output: '/api/movies'
print(parsed_url.query)    # Output: 'id=1'
```

```python
from urllib.parse import urlencode, urlunparse

params = {'search': 'python', 'page': 2}
query_string = urlencode(params)
url = urlunparse(('http', 'www.example.com', '/search', '', query_string, ''))
print(url)  # Output: 'http://www.example.com/search?search=python&page=2'
```
