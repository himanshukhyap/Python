```python
def caught_speeding(speed, is_birthday):
    if is_birthday == True:
        if 0 <= speed and speed <= 65:
            print("No ticket")
        elif 66 <= speed and speed <= 85:
            print("Small ticket")
        else:
            print("Big Ticket")
    else:
        if 0 <= speed and speed <= 60:
            print("No ticket")
        elif 61 <= speed and speed <= 80:
            print("Small ticket")
        else:
            print("Big Ticket")
```

```python
caught_speeding(81,False)
```
