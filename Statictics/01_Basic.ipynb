Data Types:

Qualitative Data (Categorical): Represents categories or labels, e.g., gender, colors, or types of animals.\
Quantitative Data (Numerical): Represents measurable quantities, e.g., height, age, or income.

Mean (Average):

Formula: The mean is calculated by summing all values and dividing by the number of values.


```python
import numpy as np
data = [10, 12, 15, 14, 13, 12, 11, 15, 16, 13]
mean = np.sum(data) / len(data)
print(mean)
meanByNumpy = np.mean(data)
print(meanByNumpy)
```
